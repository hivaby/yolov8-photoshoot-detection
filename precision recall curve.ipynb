```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터를 읽어옵니다.
data = pd.read_csv("results.csv")

# precision과 recall을 추출합니다.
precision = data.iloc[:, 4]
recall = data.iloc[:, 5]

# 그래프를 그립니다.
plt.plot(precision, color="blue", lw=2)
plt.plot(recall, color="gray", lw=2)
# 그래프의 제목과 축 라벨을 설정합니다.
plt.title("Precision-Recall Curve")
plt.xlabel("Recall")
plt.ylabel("Precision")

# 그래프를 출력합니다.
plt.show()
```
